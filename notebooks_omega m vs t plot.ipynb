```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Friedmann equation
def friedmann(t, omega_m, omega_lambda):
    H = (8 * np.pi / 3)**0.5 * (omega_m / t**2 + omega_lambda)**0.5
    return H

# Define the Ω(t)−1 function
def omega_inv(t, omega_m, omega_lambda):
    return (3 / (8 * np.pi)) * t**2 * friedmann(t, omega_m, omega_lambda)**2 - 1

# Define the time axis from t = 10^-20 seconds to the age of the universe
t = np.logspace(-20, np.log10(13.8e9), 1000)

# Define the values of omega_m and omega_lambda
omega_m = 0.3
omega_lambda = 0.7

# Calculate Ω(t)−1
omega_inv_t = omega_inv(t, omega_m, omega_lambda)

# Plot Ω(t)−1 versus time in seconds using logarithmic axes
plt.figure()
plt.loglog(t, omega_inv_t)
plt.xlabel('Time (s)')
plt.ylabel(r'$\Omega(t)^{-1}$')
plt.title('Dependence of Ω(t)−1 on other parameters')
plt.show()
```
